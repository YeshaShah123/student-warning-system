# 01 EDA
Exploratory analysis for the UCI Student Performance math dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/student-mat.csv', sep=';')
print(df.shape)
df.head()
```

```python
df['pass'] = (df['G3'] >= 10).astype(int)
print(df['pass'].value_counts(normalize=True).mul(100).round(1))
print(df.info())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='pass')
plt.title('Pass/Fail Distribution')
plt.tight_layout()
plt.show()
plt.close()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x='pass', y='absences')
plt.title('Absences by Outcome')
plt.tight_layout()
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.select_dtypes('number').corr(), cmap='vlag', center=0)
plt.title('Numeric Feature Correlation')
plt.tight_layout()
plt.show()
plt.close()
```
